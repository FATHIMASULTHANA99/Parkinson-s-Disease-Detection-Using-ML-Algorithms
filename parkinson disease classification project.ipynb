### The health status of the subjects is denoted in the 'status' column as follows: '1' indicates Parkinson's disease, while '0' represents a healthy status.

## data preprocessing

splitting data into features and target varibales

```python
x=df.drop('status', axis=True)
x.head()
```

```python
y=df['status']
y.head()
```

##  feature scaling

```python
from sklearn.preprocessing import MinMaxScaler
sc=MinMaxScaler()
```

```python
x=sc.fit_transform(x)
x
```

## Data preparation

#### we did split our dataset into independent and dependent variables

#### Now splitting  further into training and testing sets.

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print('shape of x_train; ',x_train.shape)
print('shape of x_test; ',x_test.shape)
print('shape of y_train; ',y_train.shape)
print('shape of y_test; ',y_test.shape)
```

## model building

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.svm import 'SVC'

```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
models={'LogisticRegression':LogisticRegression(),
      'GaussianNB':GaussianNB(),
      'DecisionTreeClassifier':DecisionTreeClassifier(),
      'RandomForestClassifier':RandomForestClassifier(),
      'AdaBoostClassifier':AdaBoostClassifier(),
      'GradientBoostingClassifier':GradientBoostingClassifier(),'SVC':SVC()}
```

```python
for model_name,model in models.items():
    model.fit(x_train,y_train)
    y_pred=model.predict(x_test)
    accuracy = accuracy_score(y_test, y_pred)
    cm = confusion_matrix(y_test, y_pred)
    report=classification_report(y_test,y_pred)
    print(f"{model_name} Accuracy: {accuracy:.2f}")
    print(cm)
    print(report)
```
